```python
%matplotlib inline
import math
import matplotlib.pyplot as plt
```

```python
def is_natural(N):
    if isinstance(N, int) and 1 <= N:
        return True
    else:
        return False

def get_divisor(N):
    if is_natural(N) == False:
        raise
    divisor_set = set()
    rt = int(math.sqrt(N))

    for i in range(1, rt + 1):
        if N % i == 0:
            divisor_set.add(int(i) )
            divisor_set.add(int(N/i))

    divisor_list = list(divisor_set)
    divisor_list.sort()
    return divisor_list

def divisor_function(x, N):
    divisor = get_divisor(N)
    s = 0
    for i in divisor:
        s += i**x
    return s
```

```python
def plot_divisor_function(x, N):
    x_coord = []
    y_coord= []
    
    for i in range(x, N):
        x_coord.append(i)
        y_coord.append( divisor_function(1, i) )
        
    plt.plot(x_coord, y_coord)
    plt.show()
```

```python
plot_divisor_function(1, 100)
```
